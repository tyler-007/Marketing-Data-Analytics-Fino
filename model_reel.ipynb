```python
# Define a function to classify captions

def classify_owner(row):
    caption = row['caption']
    ownerFullName = row['ownerFullName']
    
    if pd.notna(caption) and ('FinoPay' in caption or 'Fino Payments Bank' in ownerFullName):
        return 'finopay'
    elif pd.notna(caption) and ('Airtel Payments Bank' in caption or 'airtel' in caption.lower() or 'Airtel Payments Bank' in ownerFullName):
        return 'airtel'
    elif pd.notna(caption) and ('IPPB' in caption or 'India Post Payments Bank' in caption or 'India Post Payments Bank' in ownerFullName):
        return 'india post'
    else:
        return 'other'

# Apply the function to create a new column 'ownerClassification'
dataset['ownerClassification'] = dataset.apply(classify_owner, axis=1)

```

```python
airtel_data = dataset[dataset['ownerClassification'] == 'airtel']
india_post_data = dataset[dataset['ownerClassification'] == 'india post']
finopay_data = dataset[dataset['ownerClassification'] == 'finopay']
bin_edges = np.arange(0, dataset['videoDuration'].max() + 11, 10)

# Function to plot pie chart
def plot_pie_chart(data, title):
    # Create bins and count occurrences
    time_bins = pd.cut(data['videoDuration'], bins=bin_edges)
    duration_counts = time_bins.value_counts(sort=False)

    # Plot pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(duration_counts, labels=duration_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(title)
    plt.show()

# Plot pie charts for each company
plot_pie_chart(airtel_data, 'Airtel - Time Duration Distribution')
plot_pie_chart(india_post_data, 'India Post - Time Duration Distribution')
plot_pie_chart(finopay_data, 'Finopay - Time Duration Distribution')
```

```python
from collections import Counter
import re
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def generate_wordcloud(data):
    def extract_hashtags_from_captions(data):
        all_captions = ' '.join(data['caption'].astype(str))
        hashtags = re.findall(r'#\w+', all_captions)
        hashtags_counter = Counter(hashtags)
        hashtags_dict = dict(hashtags_counter)
        return hashtags_dict, all_captions

    hashtags_dict, all_captions = extract_hashtags_from_captions(data)

    # Generate WordCloud
    wordcloud = WordCloud(
        width=800,
        height=400,
        background_color='white',
        min_word_length=3,
        max_words=100,
        collocations=False
    ).generate(all_captions)

    # Plot the WordCloud image
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

#generate_wordcloud(airtel_data)
#generate_wordcloud(finopay_data)
#generate_wordcloud(india_post_data)
```

```python
from transformers import TapasTokenizer, TapasForQuestionAnswering
import pandas as pd

# Initialize the tokenizer and model
tokenizer = TapasTokenizer.from_pretrained("google/tapas-large-finetuned-wtq")
model = TapasForQuestionAnswering.from_pretrained("google/tapas-large-finetuned-wtq")

# Example table as a pandas DataFrame
data = {
    "Column 1": ["Row1", "Row2", "Row3"],
    "Column 2": ["Good", "Bad", "Ugly"]
}
table = pd.DataFrame(data)

# Define queries
queries = ["What are the positive highlights?", "What are the main complaints?"]

# Tokenize the table and queries
inputs = tokenizer(table=table, queries=queries, padding="max_length", return_tensors="pt")

# Perform question answering
outputs = model(**inputs)

# Extract and decode the answers
answers = tokenizer.batch_decode(outputs.logits.argmax(-1))
print(answers)
```
